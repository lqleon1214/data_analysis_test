> 根据震级来显示不同的颜色

```python
def get_marker_color(magnitude):
    # Returns green for small earthquakes, yellow for moderate
    #  earthquakes, and red for significant earthquakes.
    if magnitude < 3.0:
        return 'go'
    elif magnitude < 5.0:
        return 'yo'
    else:
        return 'ro'

eq_map = Basemap(projection='robin', resolution='l', area_thresh=1000.0,
                lat_0=39.9, lon_0=116.3)
eq_map.drawcoastlines()
eq_map.drawcountries()
eq_map.fillcontinents(color='gray')
eq_map.drawmapboundary()
eq_map.drawmeridians(np.arange(0, 360, 30))
eq_map.drawparallels(np.arange(-90, 90, 30))

min_marker_size = 2
for lon, lat, mag in zip(df['longitude'].tolist(), df['latitude'].tolist(), df['mag'].tolist()):
    x, y = eq_map(lon, lat)
    msize = mag * min_marker_size
    marker_string = get_marker_color(mag)
    eq_map.plot(x, y, marker_string, markersize=msize)
```

```python
def get_marker_color(magnitude):
    # Returns green for small earthquakes, yellow for moderate
    #  earthquakes, and red for significant earthquakes.
    if magnitude < 3.0:
        return 'go'
    elif magnitude < 5.0:
        return 'yo'
    else:
        return 'ro'

eq_map = Basemap(projection='robin', resolution='l', area_thresh=1000.0,
                lat_0=39.9, lon_0=116.3)
eq_map.drawcoastlines()
eq_map.drawcountries()
eq_map.fillcontinents(color='gray')
eq_map.drawmapboundary()
eq_map.drawmeridians(np.arange(0, 360, 30))
eq_map.drawparallels(np.arange(-90, 90, 30))

min_marker_size = 2
for lon, lat, mag in zip(df['longitude'].tolist(), df['latitude'].tolist(), df['mag'].tolist()):
    x, y = eq_map(lon, lat)
    msize = mag * min_marker_size
    marker_string = get_marker_color(mag)
    eq_map.plot(x, y, marker_string, markersize=msize)

title_string = 'Earthquakes of Magnitude 1.0 or Greater\n'
title_string += '%s through %s' % (df['time'][-1:].tolist()[0][:10], df['time'][0][:10])
plt.title(title_string)
```

> ### 画海洋的方式很简单，只要我们把对大陆的填充取消掉，并且调用 **`bluemarble`** 方法

```python
def get_marker_color(magnitude):
    # Returns green for small earthquakes, yellow for moderate
    #  earthquakes, and red for significant earthquakes.
    if magnitude < 3.0:
        return 'go'
    elif magnitude < 5.0:
        return 'yo'
    else:
        return 'ro'

eq_map = Basemap(projection='robin', resolution='l', area_thresh=1000.0,
                lat_0=39.9, lon_0=116.3)
eq_map.drawcoastlines()
eq_map.drawcountries()
eq_map.bluemarble()
# eq_map.fillcontinents(color='gray')
eq_map.drawmapboundary()
eq_map.drawmeridians(np.arange(0, 360, 30))
eq_map.drawparallels(np.arange(-90, 90, 30))

min_marker_size = 2
for lon, lat, mag in zip(df['longitude'].tolist(), df['latitude'].tolist(), df['mag'].tolist()):
    x, y = eq_map(lon, lat)
    msize = mag * min_marker_size
    marker_string = get_marker_color(mag)
    eq_map.plot(x, y, marker_string, markersize=msize)

title_string = 'Earthquakes of Magnitude 1.0 or Greater\n'
title_string += '%s through %s' % (df['time'][-1:].tolist()[0][:10], df['time'][0][:10])
plt.title(title_string)
```
