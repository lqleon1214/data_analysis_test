```python
import numpy as np
import pandas as pd
import scipy.stats as stats
pd.set_option('display.max_rows', 10)
pd.set_option('display.max_columns',10)
```

```python
data = pd.read_table('./data/reguser.txt')
data.head()
```

```python
data.tail()
```

```python
data['lost'] = data['reg'] - data['stay']
data.head()
```

```python
group = pd.groupby(data, by='type')
group.head()
```

```python
observed = group.agg(np.sum)
observed
```

```python
# 流失和留存人数是否相互独立
stats.chi2_contingency(observed=observed[['stay', 'lost']])     # 用户的类别的确影响了用户的次日留存情况，并不是相互独立。
```
